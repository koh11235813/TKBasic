### 人工知能 (AI) 

人間の持つ自然知能との対比となる人工知能では、人間が考えてとる行動を計算機に再現させたものです。単純に計算機が条件分岐だけを繰り返すものも広義には人工知能に含まれます。ただし、近年では、ディープラーニングを含めた機械学習のことを狭義の人工知能として呼んでいる例が多く見受けられます。  

人工知能は、汎用型人工知能と特化型人工知能の 2 つに大きく分けることができます。  

汎用型人工知能では意思を持ち、自ら考えて行動するような人工知能を目指しています。答えが与えられていない問題設定に対しても、学習によってその問題が解けるように人工知能自身が自動的に成長していきます。  
汎用型人工知能はまだ実用段階であるものが少ないのが現状です。  

そして、特化型人工知能は、ある特定の作業や問題設定に特化して学習を行い、人間の与えた行動を模倣することを目指しています。汎用性はありませんが、実用面ではこちらの開発のほうが早く進むため、今ビジネスで使われているほとんどが特化型人工知能です。本資料内での人工知能は一般的にこの特化型人工知能を指します。

### 機械学習

**機械学習 (Machine learning) **はデータからその規則性・関連性などを見つけ出し（学習）、学習したパターンを用いて新しいデータに対して、予測を行います（推論）。  

具体的にデータとは、上図に示すように画像、音声、自然言語や表データなど様々な形式のデータを指します。  
これらのデータを計算機で扱うために定量化して数値に落とし込みます。また、そこから問題設定に応じて不必要なデータを除いたり、データの特徴をより見つけやすくするような処理を施します。（特徴抽出）  

そして、そのデータの規則性・関連性を見つけだします。計算機がこの規則性・関係性を見つける方法にはアルゴリズムが用いられます。  
アルゴリズムとは、計算機にどの様な流れで処理を行わせるのかを定めた一連手順のことです。機械学習にはデータから規則性・関連性を見つけ出すためのいくつものアルゴリズムが存在します。  

ディープラーニングもその機械学習アルゴリズムのひとつであり、他にも、決定木やサポートベクトルマシン (SVM) など様々な機械学習アルゴリズムが存在しています。  

また、機械学習は大きく下記の 3 種類に分けられることも覚えておきましょう。  

- 教師あり学習
- 教師なし学習
- 強化学習

「教師あり学習」と「教師なし学習」に関しては、本講座で取り扱うので後ほど詳しく確認します。  

「[強化学習](https://ja.wikipedia.org/wiki/%E5%BC%B7%E5%8C%96%E5%AD%A6%E7%BF%92)」は AlphaGo が囲碁で人間に勝利して以来、注目を浴びるようになってきています。しかし、ゲームやロボット系の分野以外では、まだまだ実用的な技術ではないと言えるでしょう。理由としては、強化学習の考え方や実装自体はそれほど難しくないのですが、強化学習ではシミュレーター、もしくは実機が必要となり、これらを準備できる業界がゲームやロボットに限られているためです。  

### 学習と推論

実務で最もよく使用される教師あり学習の基本的な概念について確認しましょう。  

まずはじめに、機械学習を理解する上で重要な概念となる**学習 (Training) **、**推論 (Inference) **と呼ばれる 2 つのステップについて確認します。  
例として、カメラで撮影した顔画像から鈴木さん、佐藤さんのような、名前の判定を行う機械学習アルゴリズムを想定します。

#### 学習

機械学習が顔画像から名前の判定を行うためには、人間と同様にこの顔が鈴木さんで、この顔が佐藤さんで、と勉強する必要があります。  
この工程のことを学習もしくは訓練と呼びます。英語では、training であるため、直訳では訓練のほうが妥当ですが、日本語の記事などでは学習という言葉が用いられることが多いため、本講座でも学習と呼ぶこととします。

学習を行うためには**入力値**と**目標値**という 2 種類のデータが必要になります。  
下記画像の例では入力値は人の顔画像になります。そして、目標値はその人の顔画像に対応する名前になります。  
機械学習はこの入力値（情報）と目標値（答え）の組み合わせの規則性・関連性を見つけ出す事により、学習していない未知の入力値に対して予測を行うことが可能になります。  

下記画像中央の**モデル**の部分はディープラーニングなどの機械学習のアルゴリズムを指します。このモデルを入力値と目標値を与えることにより学習を行います。  

![学習](http://drive.google.com/uc?export=view&id=1m_S8vsDvA8fxy0h6G6MXjuOtXvCOpNVm)


#### 推論

学習を終えると、入力値と目標値の規則性・関連性を学習した、**学習済みモデル**が得られます。この学習済みモデルを利用して、学習していない未知の入力値に対して予測する工程を推論と呼びます。   
 
![推論](http://drive.google.com/uc?export=view&id=1UXeBClEsUBXQeRjiyKarfmbGu0BpxFuC)

この概念や単語が本資料で重要となるため、この教師あり学習の概念についてはしっかりここで理解しておきましょう。  

### ディープラーニング

ディープラーニングは機械学習アルゴリズムの 1 種ではありますが、他のアルゴリズムと比べて画像処理などの分野で大きな進展をもたらしました。  

![ディープラーニング](http://drive.google.com/uc?export=view&id=1ichxVdehHKcdFrt0NvqcexhdTo4CgHpn)

図の上側の流れがこれまでの機械学習アルゴリズムの開発の流れで、下側がディープラーニング登場後の流れです。  

たとえば、顔の画像と人の名前を関係性・規則性を見つける問題設定の場合、単純に加工前の画像を入力として与えたとしても、どこに注目して良いかを見つけることが難しく、うまく予測ができない問題がありました。そこで、特徴抽出と呼ばれる属人的に、画像を機械が認識しやすくなるように前処理を行う工程をはさみます。  

これにより、ある程度良い結果が得ることも可能です。しかし、問題設定が変わるごとに毎回特徴抽出の方法を考案する必要があり、複雑な問題設定では適用できないという問題がありました。  

その一方、ディープラーニングでは、この特徴抽出の工程をアルゴリズムの一部として組み込むことに成功し、データに基づいて特徴抽出まで学習できることで、属人的であった工程を省くことができました。この進歩により、これまで以上に複雑な問題設定に取り組めることができるようになりました。

### 企画・ヒアリング

機械学習案件に取り組む場合、まずこの「企画・ヒアリング」から始まります。まず企画を練り、どのようなシステムを実現したいかを話し合います。よくあるケースとして「どのようなシステムを作れば良いか」をまず考え、システム構成等の設計を始めます。しかし、その前にそのシステムは「なぜ必要か」そして「機械学習でないと解決できないのか」ということを考える必要があります。  

本来の目的が何かの問題解決であり、機械学習を使用することはその手段の 1 つであるという認識を持つことが重要になります。問題設定を見直してみると、既存のロジックベース（条件分岐）でも実装可能なケースであることも多々あります。このような案件に関して、仮に機械学習などのモデルを作成した場合、うまくいかない上に中がブラックボックスであるため、問題の原因究明を行うことが困難となります。また、コストもかかり、機械学習が結果的に技術的負債となる場合があります。この問題を避けるためにも、技術的負債が残らず継続的に運用可能かを踏まえた上で、企画から技術に落とし込んでいく必要があります。

### 環境構築

企画が固まり、実際に解析を行い始める前に躓く一つのポイントとして、「環境構築」があります。  

手元の PC （ローカル）で解析を行う場合は、クリック１つでインストールできるソフトウェアが配布されている便利な時代なので問題ありませんが、[GPU](https://ja.wikipedia.org/wiki/Graphics_Processing_Unit) (Graphic processing unit) を使用する機会も多く、その際は、下図に示すように、リモートにある解析用サーバーを操作するスキル、例えば、[Linux](https://ja.wikipedia.org/wiki/Linux) をコマンドで操作しながらソフトウェアをインストールするスキルが必要になります。  

また、環境を構築しただけでは終わらず、ローカルで使用できるよう、リモートとローカルの紐づけも行わなければいけません。これまでコンピュータに深く関わってこなかった人に、単独でこれらの作業は難しい作業と言えます。  

Python では、Jupyter Notebook と呼ばれる Web ブラウザベースの実行環境があります。対話側のように実行結果を都度確認でき、コードをまとめて実行できるスクリプト型の良さも取り入れており、データ解析に好んで用いられます。さらに Jupyter Notebook は、 Web ブラウザベースであるため、リモートであってもローカルであっても、環境の差をほとんど意識することなく解析を行うこと可能です。

![環境構築](http://drive.google.com/uc?export=view&id=1BeeIBKUaKx1Ef179GsvR-OQ2a1YbtckY)

Docker と呼ばれる仮想化技術を取り入れるケースも増えており、ローカルで構築した環境を OS の異なるリモートでも即時に全く同じものが移植できるようになっています。デプロイを行う際のコストを大幅に下げることができると同時に、コードのみならず環境ごと共有することができる技術です。  
環境構築において Docker の技術を理解することは重要だと言えます。  

### 目標値（教師データ）作成

環境構築が完了し、機械学習の教師あり学習に取り掛かる際には「学習と推論」の章でお伝えした通り、入力値、目標値（教師データとも呼ばれます。）のペアを準備しておく必要があります。例えば、顔写真（入力値）から人の名前（目標値）を予測するモデルを作成する際には、モデルを学習させる前に目標値となる名前を準備しておく必要があります。

物体検出と呼ばれる画像の中から物体の位置とクラス（グループ）を検出する技術では、目標値に物体の位置とクラスの答えとなるラベルを準備する必要があります。また、データの数は多ければ何千、何万というデータを準備する必要がある場合もあります。このような理由から、目標値の準備には想像以上に時間が要する場合があるため、この点に関しては「企画・ヒアリング」の段階から考慮しておき、工数などの見積もりを行う必要がある点も注意しておきましょう。

### 前処理・構造化

計算機は数値化されていないデータは取り扱うことができません。モデルを学習する際に、文章や画像などの非構造化データと呼ばれるデータをどのように機械学習の入力値として数値化（ベクトル化）して良いかわからない問題があります。そのため、データの取り扱い方法を学び、非構造化データを構造化データに変換するスキルを身につける必要があります。   

![前処理・構造化](http://drive.google.com/uc?export=view&id=1aL0fZ9Lquxq5rqSt8r3u7_J4pRFH1otj)

そして、この構造化と同じタイミングで、外れ値除去や標準化など、問題設定に応じたデータの前処理が必要となります。画像であればディープラーニングを使えば良いというものではありません。とりあえずディープラーニングを使用するも思うような精度が出ず、その原因を探りながら、結果的にデータの前処理を行っていきます。機械学習に任せればすべてがデータに基づいて解決されるわけではなく、現場での解析は、モデルを作りながら、どのような原理になっているかを人間が理解し、 そのノウハウを反映することで精度を向上させていきます。

### モデル構築

構造化されたデータに対して、入力値と目標値の規則性・関連性をモデルに学習させる段階です。実務におけるモデル構築では、scikit-learn や TensorFlow 、PyTorch　といった機械学習向けのパッケージ/フレームワークを使用することが多くあります。理由は、モデル化がうまくいかない際に、「プログラムが間違っている」か「使用する手法やデータが悪い」かを切り分けるためです。研究では、 フレームワークに頼る事なくモデル構築を行う場合も少なくありませんが、ここにはプログラムの間違いというリスクが潜んでいることを忘れてはいけません。  

例えば scikit-learn の場合、同じインターフェースで様々な手法を試すことができるため、初学者でも高度な機械学習アルゴリズムを使用することが可能です。しかし、使用可能であることと適切に使用できることは異なります。各機械学習アルゴリズムの数学を全て理解することは難易度の高いことです。しかし、ただ使用するのではなく、それぞれのアルゴリズムの概要と特性を理解し、使い分けができるようになりましょう。  

また、モデル構築に取り組む上でよく起こる問題として、予測精度何 % を要件とするのかが明確でないがゆえに、いつまでも解析を終えられないことが挙げられます。
例えば、要件が決まっていないため、精度が 80% のとき、これで運用して良いのか、まだ精度を上げる努力が必要なのかという判断を下すことができないケースです。
データ解析側からその目標となる精度を決めることはできません。もちろん、精度 100% であれば終了できます。しかし、ほとんどの問題設定では精度 100% が出ることはありません。よって、ビジネス側として「精度 100% が出せないモデルでどのように効率化するか」を先に考えておく必要があります。

精度 100% が出ないということは、下図に示すように、 人間がカバーする必要があります。このように 人工知能 を組み込んだプロセスの中に、人間がカバーするように運用することを「ヒューマン イン ザ ループ (Human in the loop) 」と呼びます。

![モデル構築](http://drive.google.com/uc?export=view&id=1gv1cPsltkS-QbzyWhkWLA9iMOPkW_XZv)

機械学習を導入するためにまず第一に、機械学習を導入すれば全てが自動化できるようなものではない、という認識を持つ必要があります。機械学習の導入までは段階を踏んで検討する必要があります。  

・手動オペレーション  
まずは手動オペレーショ ンでビジネスが回るように設計します。そのオペレーションの中で、作業手順が明確かつ費用対効果も高い作業に対して、コスト削減の手段を検討します。
  
・[RPA (Robotic process automation)](https://ja.wikipedia.org/wiki/%E3%83%AD%E3%83%9C%E3%83%86%E3%82%A3%E3%83%83%E3%82%AF%E3%83%BB%E3%83%97%E3%83%AD%E3%82%BB%E3%82%B9%E3%83%BB%E3%82%AA%E3%83%BC%E3%83%88%E3%83%A1%E3%83%BC%E3%82%B7%E3%83%A7%E3%83%B3) or [クラウドソーシング](https://ja.wikipedia.org/wiki/%E3%82%AF%E3%83%A9%E3%82%A6%E3%83%89%E3%82%BD%E3%83%BC%E3%82%B7%E3%83%B3%E3%82%B0)  
オペレーションの中でロジックベースで言語化できるものに対して、RPA、クラウドソーシングを用いて自動化あるいは外部委託します。自動化することによって、費用対効果が高い場合は RPA を、さほど高くない場合には導入がより簡単であるクラウドソーシングを活用します。機械学習と比較して、両者とも導入までの時間が短い傾向にあり、今までの言語化できる作業を自動化もしくは外注しているため、安定して成果をあげやすいです。
  
・機械学習  
RPA であれば、人間側でロジックを設計し、プログラミングを行っていきますが、画像や自然言語などロジックを言語化しにくかったり、制御構文が増えすぎてプログラミングの工数が多くなった場合に機械学習によるモデル化を検討するといいでしょう。また、業務の一部分を機械学習によって自動化し、その最終確認にクラウドソーシングを活用するといった運用も安定し、効果的だと考えられます。  

「とりあえず機械学習」ではなく、 うまく導入するための「人間のオペレーションによる冗長化」を忘れないようにしましょう。

### 仮運用・検証

手持ちの入力値と目標値を使用して構築した学習済みモデルはどの程度の予測精度があるのかを学習の段階で検証を行います。  
しかし、その検証もあくまで手持ちのデータの一部を使用して行うため、プロダクトとして本番環境で未知の入力値を使用した際に同様の予測精度が担保されているとはかぎりません。  

そのため、プロトタイプ（試作品）を作成し、実際に本番環境で仮運用を行い、本番環境の未知の入力値に対しての予測精度の検証を行う必要があります。  
モデル構築段階での検証結果のみをそのモデルの予測精度を決めてしまうことには危険性があることを覚えておきましょう。  

また、このようなプロトタイプを作成する際には Web アプリケーションに関する知識を学ぶ必要があります。

### システム統合

システム統合を行う際はクライアント側のシステムに合わせ、Web API として実装して連携、もしくはデバイスへの組み込みを行うことが多いです。

組み込み系のデバイスを使用する際には、ハードウェア側のスペックに注意が必要です。解析を行う場合は気にならなかった問題が浮上することがあります。例えば、 メモリ量の問題や計算速度です。近年注目を集めているディープラーニングでは、学習済みモデルのサイズが大きい場合が多く、時には 1GB を超える場合もります。そのため、デバイス側のメモリが足りなくなる場合があります。また、GPU を搭載しないデバイスでは、学習済みモデルを使った推論時に想定よりも時間がかかることがあります。モデル構築を行う際には、 予測精度だけでなく、最終的に組み込む先のハードウェア側のスペック、モデルのサイズや GPU なしでの計算速度まで考慮して手法の選定を行う必要があります。 

一方、組み込みデバイスのスペックでカバーするのではなく、クラウド上にある GPU も含んだ高スペックな計算機で推論を行わせ、その結果をデバイス側で使用するといったクラウドとデバイスという関係で実装するための環境も整ってきています。この方法であれば、デバイス側のスペックに依存せずにモデルの実装を行うことができます。ただし、インターネット通信できることが前提であるため、通信環境の有無に注意が必要です。

この `pip3` とは、Python の中で使用するパッケージと呼ばれる便利なツールを管理してくれるためのツールです。

#### brew と pip の違い

`brew` も `pip3` も何かをインストールするために使うものですが「**Mac 内で使用するソフトウェアのインストールには brew**」を使用し、「**Python 内で使用するパッケージのインストールには pip**」を使用します。

`brew`が Mac 内で使用する `python3` や `pip3` などのソフトウェアを管理し、`pip3`がその`python`で使用するためのパッケージを管理するといった構造になっています。

```
Homebrew
 |- python
 |- pip3
     |- Pythonで使用するパッケージ...
 |- etc...
```

そのため解析を行う際に他のパッケージをインストールする際には基本的には `pip3` を用います。

#### Python のバージョンを確認

Python をインストールしたのですが、その時々によって、インストールされるバージョンが変わります（基本的には最新のバージョンをインストールするため）。

インストールしたバージョンを確認しておきましょう。

```bash

 python3 --version

```

帰ってきたバージョン情報の数字が `3` から始まっていれば問題ありません。
